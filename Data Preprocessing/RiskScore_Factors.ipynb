```python
import numpy as np
import pandas as pd
import random
```

```python
df1=pd.read_excel("Locations and Headcount.xlsx")
```

```python
num_of_offices=df1.groupby("Geo Country").count()["Area"].tolist()
```

```python
headcount=df1.groupby("Geo Country").sum()
```

```python
randomlist = []
for i in range(0,151):
    n = random.randint(0,10)
    randomlist.append(n)
headcount["Number_of_office"]=num_of_offices
headcount["Number_of_incident"]=randomlist
```

```python
randomlist = []
for i in range(0,151):
    n = random.randint(300,520)
    randomlist.append(n)
headcount["Client_Presence"]=randomlist
```

```python
def preprocessing_incident(cols):
    if(cols[0]>100000):
        cols[1]=int(cols[1]*0.5)
    elif(cols[0]>70000):
        cols[1]=int(cols[1]*0.6)
    elif(cols[0]>30000):
        cols[1]=int(cols[1]*0.7)
    elif(cols[0]>10000):
         cols[1]=int(cols[1]*0.8)
    elif(cols[0]>5000):
         cols[1]=int(cols[1]*0.9)
    if(cols[1]>7):
        return 5
    elif(cols[1]>5):
        return 4
    elif(cols[1]>4):
        return 3
    elif(cols[1]>2):
        return 2
    else:
        return 1
def preprocessing_client(cols):
    if(cols[0]>100000):
        cols[1]=int(cols[1]*0.6)
    elif(cols[0]>70000):
        cols[1]=int(cols[1]*0.7)
    elif(cols[0]>30000):
        cols[1]=int(cols[1]*0.8)
    elif(cols[0]>5000):
         cols[1]=int(cols[1]*0.9)
    if(cols[1]>475):
        return 5
    elif(cols[1]>450):
        return 4
    elif(cols[1]>410):
        return 3
    elif(cols[1]>380):
        return 2
    else:
        return 1
def preprocessing_office(cols):
    if(cols[1]>7):
        return 5
    elif(cols[1]>5):
        return 4
    elif(cols[1]>4):
        return 3
    elif(cols[1]>2):
        return 2
    else:
        return 1
def preprocessing_headcount(cols):
    if(cols[0]>100000):
        cols[0]=int(cols[0]*0.4)
    elif(cols[0]>70000):
        cols[0]=int(cols[0]*0.5)
    elif(cols[0]>30000):
        cols[0]=int(cols[0]*0.6)
    elif(cols[0]>5000):
         cols[0]=int(cols[0]*0.7)
    if(cols[0]>5000):
        return 5
    elif(cols[0]>3000):
        return 4
    elif(cols[0]>1000):
        return 3
    elif(cols[0]>500):
        return 2
    else:
        return 1
```

```python
headcount.head(5)
```

```python
headcount['Number_of_incident'] = headcount[['Head Count','Number_of_incident']].apply(preprocessing_incident,axis=1)
headcount['Client_Presence'] = headcount[['Head Count','Client_Presence']].apply(preprocessing_client,axis=1)
headcount['Number_of_office'] = headcount[['Head Count','Number_of_office']].apply(preprocessing_office,axis=1)
headcount['Head Count'] = headcount[['Head Count','Number_of_incident']].apply(preprocessing_headcount,axis=1)
```
